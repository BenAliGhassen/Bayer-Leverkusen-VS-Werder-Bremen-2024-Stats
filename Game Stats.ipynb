```python
parser = Sbopen()

match = sb.events(match_id=3895302)

df, refered, freeze, tactics = parser.event(3895302)
```

```python
df= df[df['team_name'] == 'Bayer Leverkusen']

df.head()
```

```python
df.columns
```

```python
subs = df[df['type_name'] == 'Substitution']


firstSub = subs['minute'].min()

firstSub
```

```python
df_passes = df[['id','player_name','player_id','team_name','pass_recipient_id','x','y','end_x','end_y','pass_recipient_name','outcome_id','outcome_name','minute']]
df_shots = df[['id','player_id','team_name','shot_statsbomb_xg','outcome_name','x', 'y']]
xahaka_passes = df_passes[df_passes['player_name'] == 'Granit Xhaka']
df_passes.head()
```

```python
df_passes = df_passes[df_passes['outcome_name'].isnull()]

df_passes.head()
```

```python
df_passes.count()
```

```python
df_passes = df_passes[df_passes['minute'] < firstSub]

df_passes.count()
```

```python
df_lineup = parser.lineup(3895302)

jersyNb = df_lineup[['player_id','jersey_number']]

jersyNb
```

```python
df_passes = pd.merge(df_passes , jersyNb, on='player_id')

df_passes.rename(columns={'jersey_number' : 'passer'}, inplace= True)

df_passes
```

```python

jersyNb.rename(columns={'player_id' : 'pass_recipient_id'}, inplace=True)
df_passes = pd.merge(df_passes , jersyNb, on='pass_recipient_id')

df_passes.rename(columns={'jersey_number' : 'recipient'}, inplace= True)

df_passes
```

```python
avg_location = df_passes.groupby('passer').agg({'x' :['mean'] , 'y' : ['mean','count']})

avg_location.columns = ['x','y','count']

avg_location
```

```python
df_passes = df_passes.groupby(['passer','recipient']).id.count().reset_index()

df_passes.rename(columns={'id': 'pass_count'},inplace=True)

df_passes.head()
```

```python
df_passes = pd.merge(df_passes , avg_location, on='passer')
avg_location = avg_location.rename_axis('recipient')
df_passes = pd.merge(df_passes , avg_location, on='recipient',suffixes=['','_end'])
df_passes
```

```python
df_passes = df_passes[df_passes['pass_count'] > 3]

df_passes.head()
```

```python
pitch = Pitch(pitch_color='white')

fig ,ax = pitch.draw()

pitch.lines(1.2* df_passes.x, .8* df_passes.y , 1.2* df_passes.x_end, .8* df_passes.y_end ,lw= df_passes.pass_count * 0.5 , color='red',ax=ax)
pitch.scatter(1.2 * avg_location.x , 0.8 * avg_location.y, s= 20 * avg_location['count'].values , color='white',edgecolor='black',ax=ax)

for index, row in avg_location.iterrows() : 
    pitch.annotate(index, xy=(1.2*row.x , .8* row.y),c='#161A30',fontweight='bold',va='center',ha='center',size=8 , ax=ax)


ax.text(
    x= 18,
    y= 0,
    s='Bayer Leverkusen Passing Network',
    fontsize = 15,
    color= 'red'
)

plt.show()    
```
